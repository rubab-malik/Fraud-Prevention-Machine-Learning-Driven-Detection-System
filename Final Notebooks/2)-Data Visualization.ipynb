## 3) Visualize the Data relationship between  features

 ### Note:
**visualize the distribution of the target variable, isFraud, in the train_transaction dataset.

#### Step 1: Distribution of the target variable 'isFraud'

```python
# Distribution of the target variable 'isFraud'
sns.countplot(x='isFraud', data=clean_train_data)
plt.title('Distribution of Fraud Transactions')
# Save the plot as an image file in the 'images' folder
plt.savefig('images/distribution_of_fraud_transactions.png', dpi=300)
plt.show()
```

#### Step 2: visualize the distribution of the TransactionAmt variable for both fraud and non-fraud transactions:

```python
# Distribution of transaction amounts for fraud and non-fraud transactions
plt.figure(figsize=(10, 5))
sns.histplot(clean_train_data[clean_train_data['isFraud'] == 0]['TransactionAmt'], kde=False, bins=50, color='blue', label='Non-fraud')
sns.histplot(clean_train_data[clean_train_data['isFraud'] == 1]['TransactionAmt'], kde=False, bins=50, color='red', label='Fraud')
plt.legend()
plt.title('Transaction Amount Distribution for Fraud and Non-Fraud Transactions')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.savefig('images/distribution of the variable for both fraud and non-fraud.png', dpi=300)
plt.show()
```

####  Step 3: heatmap to visualize the correlation between some of the numerical features in the train_transaction dataset:

```python
# Select a subset of numerical features for the correlation heatmap
numerical_features = ['TransactionAmt', 'card1', 'card2', 'card3', 'card5', 'addr1', 'addr2']

# Calculate the correlation matrix
corr_matrix = clean_train_data[numerical_features].corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.savefig('images/heatmap to visualize the correlation .png', dpi=300)
plt.show()
```
